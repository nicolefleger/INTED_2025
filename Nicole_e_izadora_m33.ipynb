```python
# Senha pr√©-definida
SENHA_CORRETA = "programacao123"
MAX_TENTATIVAS = 3

print("üîí Sistema de Login")
print(f"Voc√™ tem {MAX_TENTATIVAS} tentativas.")

for tentativa in range(1, MAX_TENTATIVAS + 1):
    senha_digitada = input(f"Tentativa {tentativa}/{MAX_TENTATIVAS}: Digite a senha: ")

    if senha_digitada == SENHA_CORRETA:
        print("\n‚úÖ Login bem-sucedido! Bem-vindo(a).")
        # Usa 'break' para sair do la√ßo imediatamente
        break
    else:
        print("‚ùå Senha incorreta.")

# A cl√°usula 'else' do 'for' √© executada apenas se o la√ßo terminar naturalmente (sem 'break')
else:
    print("\nüõë N√∫mero m√°ximo de tentativas atingido. Acesso bloqueado.")
```

```python
total_positivo = 0
TOTAL_NUMEROS = 10

print("üî¢ Soma de Valores Positivos")

for i in range(TOTAL_NUMEROS):
    try:
        numero = float(input(f"Digite o {i + 1}¬∫ n√∫mero: "))
    except ValueError:
        print("Entrada inv√°lida. Digite um n√∫mero.")
        # Opcional: para garantir que 10 n√∫meros v√°lidos sejam lidos, voc√™ pode decrementar 'i' ou ajustar a l√≥gica.
        # Simplificando, neste caso, o 'continue' pula apenas a soma.
        continue

    if numero < 0:
        print(f"N√∫mero negativo ({numero}) ignorado. ‚è≠Ô∏è")
        # Usa 'continue' para pular a linha 'total_positivo += numero'
        continue

    # Esta linha s√≥ √© executada se o n√∫mero for positivo ou zero
    total_positivo += numero

print(f"\n‚ú® A soma de todos os n√∫meros positivos digitados √©: **{total_positivo}**")56
```

```python
soma_acumulada = 0
numero_atual = 0
LIMITE_SOMA = 200

print(f"üöÄ Acumulador at√© que a soma ultrapasse {LIMITE_SOMA}")

while True:
    numero_atual += 1
    soma_acumulada += numero_atual

    print(f"Adicionando {numero_atual}. Soma atual: {soma_acumulada}")

    if soma_acumulada > LIMITE_SOMA:
        print(f"\nüõë A soma ultrapassou **{LIMITE_SOMA}**! Valor final: **{soma_acumulada}**.")
        # Usa 'break' para sair do la√ßo 'while True'
        break

print(f"O la√ßo encerrou ao somar at√© o n√∫mero **{numero_atual}**.")
```

```python
contador_palavras = 0
palavra = "" # Inicializa para entrar no la√ßo 'while'

print('üìù Contador de Palavras')
print('Digite palavras e pressione Enter. Digite "**fim**" para encerrar.')

# O la√ßo continua enquanto a palavra n√£o for "fim"
while palavra.lower() != "fim":
    palavra = input("Digite uma palavra: ")

    # Verifica se a palavra √© a de parada
    if palavra.lower() == "fim":
        # N√£o faz nada, o la√ßo vai parar na pr√≥xima checagem
        pass
    # Verifica se a palavra n√£o est√° vazia (caso o usu√°rio aperte Enter sem digitar nada)
    elif palavra.strip() != "":
        contador_palavras += 1
        print("Palavra v√°lida registrada.")
    else:
        print("Entrada vazia ignorada.")


print(f"\nüéâ Processo encerrado. Voc√™ digitou um total de **{contador_palavras}** palavras v√°lidas.")
```
