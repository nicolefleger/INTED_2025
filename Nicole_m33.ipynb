```python
maior = ""
for i in range(4):
    tarefa = input(f"Tarefa {i+1}: ")
    if len(tarefa) > len(maior):
        maior = tarefa

print("Tarefa com mais caracteres:", maior)
```

```python
pos = neg = zero = 0

for i in range(10):
    n = float(input(f"Número {i+1}: "))
    if n > 0:
        pos += 1
    elif n < 0:
        neg += 1
    else:
        zero += 1

print("Positivos:", pos)
print("Negativos:", neg)
print("Zeros:", zero)
```
